```python
from keras.models import load_model  #to load model
from time import sleep  #for time control
from keras.preprocessing.image import img_to_array #conversion
from keras.preprocessing import image 
import cv2 #computer vision
import numpy as np #numpy
import time 
# import pandas as pd
from statistics import mode
from playsound import playsound #for song
import os #files
import random
```

```python
face_classifier = cv2.CascadeClassifier(r'C:\Users\91790\Downloads\ML\Project\Facial-Expressions-Recognition-master\Facial-Expressions-Recognition-master\haarcascade_frontalface_default.xml')
classifier =load_model(r'C:\Users\91790\Downloads\ML\Project\Emotion_little_vgg2.h5')
```

```python
class_labels = ["Angry","Disgust","Fear","Happy","Neutral","Sad","Surprise"]
```

```python
cap = cv2.VideoCapture(0)

t = time.time() + 60 #for 2 mins
emotion = []

while True:
    
    _, frame = cap.read()
    labels = []
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray,1.3,5)

    
    
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h,x:x+w]
        roi_gray = cv2.resize(roi_gray,(48,48),interpolation=cv2.INTER_AREA)
    # rect,face,image = face_detector(frame)


        if np.sum([roi_gray])!=0:
            roi = roi_gray.astype('float')/255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi,axis=0)

        # make a prediction on the ROI, then lookup the class
        
            
            
            
            preds = classifier.predict(roi)[0]
            label=class_labels[preds.argmax()]
            label_position = (x,y)
            cv2.putText(frame,label,label_position,cv2.FONT_ITALIC,2,(0,0,255),3)\
            
            emotion.append(str(label))
                
#             emotion.insert(-1,'Emotions',)
            
#         if time.time()>t:
#             break
        else:
            cv2.putText(frame,'No Face Found',(20,60),cv2.FONT_HERSHEY_TRIPLEX,2,(0,0,255),3)
    
    if time.time()>t:
        break
    
    cv2.imshow('Emotion Detector',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

final_emotion = mode(emotion)

songs_path = r'C:\Users\91790\Downloads\ML\Project\Songs/' + str(final_emotion)

# print()

songs = os.listdir(songs_path)
song = random.choice(songs)

playsound(songs_path + str("/") + song)
```
